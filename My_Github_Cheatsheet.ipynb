------ git diff <br>
git diff  <br>

git diff \<directory\>  <br>
git diff -r HEAD  <br>
git diff ID1..ID2    // IDs are unique commit IDs or their first few digits  <br>
git diff HEAD..HEAD\~1  // difference between most recent and second most recent commit  <br>
<br>

git log <br>

------ git add   // To add the files to the staging area (put documents in the envelope equivalent) <br>
// We have to add a file at least once for version control to track it. ***git status*** would tell you all the untracked files in the given directory. <br>
<br>
git add \<filename\>  <br>
git add .   <br>

<br>
------ git reset   <br>
git reset HEAD \<filename\>    // To unstage the files from staging area   <br>

<br>
------git commit    <br>
git commit
git commit -m "Message here"   <br>
git commit --amend - m "new message"   <br>

<br>
----------- Commit IDs - Hash - a unique identifier for every commit <br>
40-character hexadecimal string like 7c35a3ce607a14953f070f0f83b5d74c2296ef93  <br>
It is generated by running the changes through a pseudo-random number generator called a hash function <br>


<br>
----------- git show  ()  <br>
git show   // shows all the changes made in all the commits  <br>
git show \<first few characters of hash of that commit\>   // shows changes made in that exact commit<br>
git show HEAD     // shows changes made in the most recent commit  <br>
git show HEAD\~1  // shows changes made in the second most recent commit <br>
git show HEAD\~2  // third most recent commit and so on <br>


<br>
--------- git annotate <br>
git annotate \<filename\>   //shows who made the last change to each line of a file and when   <br>  



<br>
--------- .gitignore    // to ignore files with patterns (temp/intermediate files)   <br>

// create a .gitignore file and put all the patterns that you want to ignore including wildcard ones. For example if the file contains following text <br>

<br>
build <br>
*.mpl <br>
<br>
then Git will ignore any file or directory called build (and, if it's a directory, anything in it), as well as any file whose name ends in .mpl <br>


<br>
-------- git clean  // to remove unwanted files  <br>
git clean -n    // will show you a list of files that are in the repository, but whose history Git is not currently tracking <br>
// You can alternately add those files to the staging area if you want to keep those files in track  <br>

git clean -f  // to remove the files that ***git clean -n*** showed before <br>

--------- How can I see how Git is configured?  <br>
git config --list  <br> 
// following flags can be added to the above command <br>
<br>--system   // : settings for every user on this computer.
<br>--global   // : setting for every one of your projects
<br>--local    // : setting for one specific project

<br>
